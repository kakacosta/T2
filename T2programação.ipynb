# Análise dos municipios 
Com os dados disponíveis no site da IGM (http://www.igm.cfa.org.br/downloads), Índice de Governança dos Municípios, obtive algumas análises nas áreas de demografia, finanças e violência.
Essas análises foram feitas na linguagem python e utilizando as bibliotecas pandas.


```python
# Importando o pandas 
# Criação do DataFrame com a aba Ranking
# Visualizar 
import pandas as pd
df = pd.read_excel('dados_IGM-CFA_2017.xlsx', sheet_name ='Ranking', skiprows = 1,usecols = [3,4,5,7,8,9,])
df.head()
```

```python
# contar cidades captais e interiores
df['Capital (1=Sim;0=Não)'].value_counts()
```

```python
# Criar um data frame com cidades do interior
# Cidades do interior mais populosas 
df2 = df[df['Capital (1=Sim;0=Não)'] == 0]
df2.nlargest(10,"População 2016 (IBGE)")
```

```python
# qual o estado com mais municipios? e o com menos?
df['Estado'].value_counts()
```

```python
# Quais as cidades mais populosas do país?
df.nlargest(5, "População 2016 (IBGE)")
```

```python
# Quais as cidades menos populosas do país?
df.nsmallest(5, "População 2016 (IBGE)")
```

```python
# Quais as cidades com maior área?
df.nlargest(5,'Área da unidade territorial 2015 (km²) (IBGE Cidades)' )
```

```python
# Quais as menores cidades do país?
df.nsmallest(5,'Área da unidade territorial 2015 (km²) (IBGE Cidades)' )
```

```python
# Criar um Dataframe apenas com as capitais, df3
# Quais as captais com maior área?
df3 = df[df['Capital (1=Sim;0=Não)'] == 1]
df3.nlargest(5,"Área da unidade territorial 2015 (km²) (IBGE Cidades)")
```

```python
# Quais as capitais com menor área?
df3.nsmallest(5,"Área da unidade territorial 2015 (km²) (IBGE Cidades)")
```

```python
# Criar um novo dataframe para visualisar a aba Fiscal

dfiscal = pd.read_excel('dados_IGM-CFA_2017.xlsx', sheet_name ='Fiscal', skiprows = 1, usecols=[3,4,5,12,13])
dfiscal.head()
```

```python
# Cidades com maior despesa em educação
dfiscal.nlargest(10,'Despesas com educação (STN 2016)')
```

```python
# cidades com menor despesa na educação
dfiscal.nsmallest(5,'Despesas com educação (STN 2016)')
```

```python
# Cidades com maior despesa na saúde
dfiscal.nlargest(5,'Despesas com saúde (STN 2016)')
```

```python
# Cidades com menores despesas para saúde
dfiscal.nsmallest(5,'Despesas com saúde (STN 2016)')
```

```python
# Criando um Dataframe com as cidades do interior
# Cidades do interior com maior despesa na educação
dfiscal2 = dfiscal[dfiscal['Capital (1=Sim;0=Não)'] == 0]
dfiscal2.nlargest(10,'Despesas com educação (STN 2016)')
```

```python
# Cidades do interior com maiores despesas na saúde 
dfiscal2.nlargest(10,'Despesas com saúde (STN 2016)')
```

```python
# criação de um DataFrame para utilizar apenas capitais
# capitais com menores despesas na saúde
dfiscal3 = dfiscal[dfiscal['Capital (1=Sim;0=Não)'] == 1]
dfiscal3.nsmallest(5,'Despesas com saúde (STN 2016)')
```

```python
# Capitais com menores despesas na educação
dfiscal3.nsmallest(5,'Despesas com educação (STN 2016)')
```

```python
# Criando mais um DataFrame para visualisaçãodo desempenho 
# Criando um Dataframe para capitais 
# Criando um Dataframe para interiores
desempenhof = pd.read_excel('dados_IGM-CFA_2017.xlsx', sheet_name ='Desempenho', skiprows = 1, usecols=[3,4,5,58,63,65,66])
desempenhof2 = desempenhof[desempenhof['Capital (1=Sim;0=Não)'] == 0]
desempenhof3 = desempenhof[desempenhof['Capital (1=Sim;0=Não)'] == 1]
desempenhof.head()
```

```python
# Cidades que mais ocorreram mortes por arma de fogo
desempenhof.nlargest(10, 'Total Mortes por armas de fogo (DataSUS 2015)')
```

```python
# Cidades do interior que mais ocorreram mortes por arma de fogo
desempenhof2.nlargest(10, 'Total Mortes por armas de fogo (DataSUS 2015)')
```

```python
# Capitais que menos ocorreram mortes por arma de fogo
desempenhof3.nsmallest(10, 'Total Mortes por armas de fogo (DataSUS 2015)')
```

```python
# Cidades que mais ocorreram mortes no trânsito
desempenhof.nlargest(10, 'Total Mortes no Trânsito (DataSUS 2015)')
```

```python
# Capitis com menos mortes no transito
desempenhof3.nsmallest(10, 'Total Mortes no Trânsito (DataSUS 2015)')
```

```python
# Cidades com maior nível de mortes por armas de fogo por 100.000 habitantes
desempenhof.nlargest(10, 'Mortes por armas de fogo por 100.000 habitantes (DataSUS 2015/IBGE 2015 x 100.000)')
```

```python
# Capitais com maior nível de mortes por armas de fogo por 100.000 habitantes
desempenhof3.nlargest(10, 'Mortes por armas de fogo por 100.000 habitantes (DataSUS 2015/IBGE 2015 x 100.000)')
```
